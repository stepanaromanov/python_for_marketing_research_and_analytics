## 2. Certifications Table (Grouped by Domain)

```python
# ── Colour palette per group ───────────────────────────────────────────────────
group_colors = {
    "Machine Learning & MLOps":    "#0072B2",
    "Statistics & Probability":    "#009E73",
    "Data Engineering":            "#56B4E9",
    "Marketing Analytics":         "#D55E00",
    "Revenue Operations & CRM":    "#CC79A7",
    "Web & Product Analytics":     "#F0E442",
    "Software Engineering":        "#E69F00",
}

def make_link(row):
    """Wrap certificate name in an HTML anchor if a URL exists."""
    if row["Verify URL"]:
        return f'<a href="{row["Verify URL"]}" target="_blank">{row["Certificate"]}</a>'
    return row["Certificate"]

display_df = df.copy()
display_df["Certificate"] = display_df.apply(make_link, axis=1)
display_df = display_df.drop(columns=["Verify URL"])

# ── Build styled HTML table ────────────────────────────────────────────────────
rows_html = ""
prev_group = None

for _, row in display_df.iterrows():
    group = row["Group"]
    color = group_colors.get(group, "#f9f9f9")

    if group != prev_group:
        group_count = len(df[df["Group"] == group])
        rows_html += f"""
        <tr>
          <td rowspan="{group_count}"
              style="background:{color}; font-weight:700; font-size:13px;
                     vertical-align:middle; text-align:center;
                     border-right:2px solid #ccc; padding:10px 14px;
                     white-space:nowrap;">
            {group}
          </td>
        """
        prev_group = group
    else:
        rows_html += "<tr>"

    rows_html += f"""
          <td style="padding:8px 12px;">{row['Certificate']}</td>
          <td style="padding:8px 12px; white-space:nowrap;">{row['Issuer']}</td>
          <td style="padding:8px 12px; text-align:center;">{row['Year']}</td>
          <td style="padding:8px 12px; text-align:center; font-weight:600;">{row['Est. Hours']} h</td>
          <td style="padding:8px 12px; font-size:12.5px;">{row['Core Content']}</td>
          <td style="padding:8px 12px; font-size:12.5px;">{row['Business Value']}</td>
        </tr>
    """

table_html = f"""
<style>
  table.cert-table {{
    border-collapse: collapse;
    width: 100%;
    font-family: 'Segoe UI', Arial, sans-serif;
    font-size: 13px;
    box-shadow: 0 2px 8px rgba(0,0,0,0.08);
  }}
  table.cert-table th {{
    background: #2c3e50;
    color: #ffffff;
    padding: 11px 14px;
    text-align: left;
    font-size: 12.5px;
    letter-spacing: 0.4px;
  }}
  table.cert-table td {{
    border: 1px solid #e0e0e0;
    vertical-align: top;
    line-height: 1.5;
  }}
  table.cert-table tr:hover td {{
    background: #f0f4ff !important;
  }}
  table.cert-table a {{
    color: #1a73e8;
    text-decoration: none;
    font-weight: 600;
  }}
  table.cert-table a:hover {{ text-decoration: underline; }}
</style>
<table class="cert-table">
  <thead>
    <tr>
      <th>Domain Group</th>
      <th>Certificate</th>
      <th>Issuer</th>
      <th>Year</th>
      <th>Est. Hours</th>
      <th>Core Content</th>
      <th>Business Value</th>
    </tr>
  </thead>
  <tbody>
    {rows_html}
  </tbody>
</table>
"""

```

## 3. Summary Statistics

```python
summary = (
    df.groupby("Group")
    .agg(
        Certifications=("Certificate", "count"),
        Total_Hours=("Est. Hours", "sum"),
        Avg_Hours=("Est. Hours", "mean"),
        Latest_Year=("Year", "max")
    )
    .reset_index()
    .sort_values("Total_Hours", ascending=False)
)

summary["Avg_Hours"] = summary["Avg_Hours"].round(1)

summary_style = summary.style \
    .set_caption("📊 Summary by Domain Group") \
    .background_gradient(subset=["Total_Hours"], cmap="Blues") \
    .background_gradient(subset=["Certifications"], cmap="Greens") \
    .format({"Avg_Hours": "{:.1f} h", "Total_Hours": "{} h"}) \
    .set_table_styles([{
        'selector': 'caption',
        'props': [('font-size', '15px'), ('font-weight', 'bold'), ('padding-bottom', '8px')]
    }])

display(summary_style)
print(f"\n── Grand Total ──")
print(f"  Certifications : {df['Certificate'].nunique()}")
print(f"  Learning Hours : {df['Est. Hours'].sum()} h")
print(f"  Year Range     : {df['Year'].min()} – {df['Year'].max()}")
```

## 4. Visualisations

```python
palette = list(group_colors.values())
groups  = summary["Group"].tolist()
colors  = [group_colors.get(g, "#cccccc") for g in groups]
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle("Professional Development – Overview", fontsize=15, fontweight="bold", y=1.01)

# ── Chart 1: Hours per group (horizontal bar) ─────────────────────────────────
ax1 = axes[0]
bars = ax1.barh(summary["Group"], summary["Total_Hours"], color=colors, edgecolor="white", height=0.6)
for bar, val in zip(bars, summary["Total_Hours"]):
    ax1.text(bar.get_width() + 1, bar.get_y() + bar.get_height()/2,
             f"{val} h", va="center", fontsize=10)
ax1.set_title("Total Learning Hours by Domain", fontweight="medium", pad=12)
ax1.set_xlabel("Hours")
ax1.invert_yaxis()
ax1.spines[["top", "right"]].set_visible(False)
ax1.set_xlim(0, summary["Total_Hours"].max() * 1.2)

# ── Chart 2: Certifications count pie ────────────────────────────────────────
ax2 = axes[1]
wedges, texts, autotexts = ax2.pie(
    summary["Certifications"],
    labels=None,
    colors=colors,
    autopct=lambda p: f"{p:.0f}%" if p > 6 else "",
    startangle=140,
    pctdistance=0.78,
    center=(0, 0.35), 
    radius=0.75,                        
    wedgeprops=dict(edgecolor="white", linewidth=1.5)
)
for at in autotexts:
    at.set_fontsize(9)
ax2.legend(wedges, summary["Group"], loc="lower center",
           bbox_to_anchor=(0.5, -0.1), fontsize=8, ncol=1)

# ── Chart 3: Timeline – certs per year ───────────────────────────────────────
ax3 = axes[2]
year_counts = df.groupby("Year")["Certificate"].count().reset_index()
year_hours  = df.groupby("Year")["Est. Hours"].sum().reset_index()

ax3.bar(year_hours["Year"], year_hours["Est. Hours"],
        color="#0072B2", edgecolor="white", label="Hours", zorder=3)

ax3b = ax3.twinx()
ax3b.plot(year_counts["Year"], year_counts["Certificate"],
          color="#D55E00", marker="o", linewidth=3, label="# Certs", zorder=4)

ax3.set_title("Certifications & Hours by Year", fontweight="medium", pad=12)
ax3.set_xlabel("Year")
ax3.set_ylabel("Est. Hours", color="#0072B2")
ax3.tick_params(axis="y", labelcolor="#0072B2")
ax3b.set_ylabel("Number of Certifications", color="#D55E00")
ax3b.tick_params(axis="y", labelcolor="#D55E00")
ax3b.yaxis.set_major_locator(MaxNLocator(integer=True))
ax3b.grid(False)
ax3.set_xticks(year_counts["Year"])
ax3.set_axisbelow(True)
ax3.grid(axis="y", alpha=0.5, zorder=0)
ax3.spines[["top"]].set_visible(False)

lines1, labels1 = ax3.get_legend_handles_labels()
lines2, labels2 = ax3b.get_legend_handles_labels()
ax3.legend(lines1 + lines2, labels1 + labels2, loc="upper left", fontsize=9)

plt.tight_layout()

# shift pie chart up after tight layout
shift = 0.07
pos = ax2.get_position()
ax2.set_position([pos.x0, pos.y0 + shift, pos.width, pos.height])
ax2.set_title("Share of Certifications by Domain", fontweight="medium", pad=12, y=pos.height+0.15)

plt.show()
```
